## Identificando picos de buscas do termo "Bia Haddad" ao longo do último ano

Bia Haddad Maia é uma tenista brasileira que se destacou no circuito mundial de tênis WTA no último ano. Em 7 de julho, Bia Haddad conquistou uma vitória sobre Ons Jabeur, que era a número 07 do ranking na época. O objetivo deste código é investigar e analisar o interesse dos brasileiros por Bia Haddad usando dados do Google Trends.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pytrends.request import TrendReq

# Criar objeto TrendReq.  Português e ajuste da time zone
pytrends = TrendReq(hl='pt-BR', tz=360)

# Configurar os termos de pesquisa e período de tempo
termos = ['Bia Haddad', 'Novak Djokovic', 'Carlos Alcaraz']

# Construir o payload de solicitação
pytrends.build_payload(termos, timeframe = '2022-06-06 2023-06-07', geo = 'BR')

# Obter os dados dos interesses por região
# interest_over_time() retorna uma série de tempo diária
dados = pytrends.interest_over_time()

# Plota o gráfico
dados.plot(figsize=(10, 6))
plt.xlabel('Data')
plt.ylabel('Interesse no último ano')
plt.title('Buscas no Google Trends por Bia Haddad, Carlos Alcaraz e Novak Djocovic no Brasil')
plt.show()
```

```python
from pytrends.request import TrendReq

# Cria uma instância do TrendReq
pytrends = TrendReq(hl='pt-BR', tz=360)

# Define o termo de pesquisa
termo_pesquisa = 'Bia Haddad'

# Configura os parâmetros da pesquisa
pytrends.build_payload([termo_pesquisa], timeframe='today 1-m', geo='BR')

# Obtém os dados do Google Trends para o Brasil
dados_brasil = pytrends.interest_over_time()

# Calcula a variação percentual das pesquisas
var_percentual = (dados_brasil[termo_pesquisa][-1] - dados_brasil[termo_pesquisa][0]) / dados_brasil[termo_pesquisa][0] * 100

print(f"Variação percentual das pesquisas por {termo_pesquisa} no Brasil nos último mês: {var_percentual:.2f}%")
```
